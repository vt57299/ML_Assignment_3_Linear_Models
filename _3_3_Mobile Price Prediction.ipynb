```python
import pandas as pd
```

```python
df = pd.read_csv('mobile_price_data.csv')
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True, drop = True)
```

# Converting text data to numerical

```python
df['ram_(GB)'] = pd.to_numeric(df['ram'].str.replace('GB',''))
df['battery_power_(mAh)'] = pd.to_numeric(df['battery_power'].str.replace(' mAh',''))
df['mob_width_(mm)'] = pd.to_numeric(df['mob_width'].str.replace(' mm',''))
df['mob_height_(mm)'] = pd.to_numeric(df['mob_height'].str.replace(' mm',''))
df['mob_depth_(mm)'] = pd.to_numeric(df['mob_depth'].str.replace(' mm',''))
df['mob_weight_(g)'] = pd.to_numeric(df['mob_weight'].str.replace(' g',''))
df['mp_speed_(GHz)'] = pd.to_numeric(df['mp_speed'].str.replace(' GHz',''))
df['int_memory(GB)'] = pd.to_numeric(df['int_memory'].str.replace(' GB',''))
df['disp_size_(cm)'] = pd.to_numeric(df['disp_size'].str.split(' ').str[0])
df['mobile_price'] = pd.to_numeric(df['mobile_price'].str.replace(r'[^\d.]', '', regex=True))
# df['os'] = ((pd.Series(df['os'])).str.extract(r'(\d+(?:\.\d+(?:\.\d+)?)?)', expand=False)).astype(float)
```

```python
df['os'] = df['os'].replace({'Android Pie 9.0':'Android Pie 9','Android Pie 10':'Android 10','Android Oreo 8.0':'Android Oreo 8','Android Oreo 8.1.0':'Android Oreo 8.1'})
```

```python
cat_cols = ['mobile_name','os','mobile_color','resolution','p_cam','f_cam','network']
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for cols in cat_cols:
    if df[cols].dtype == object:
        df[cols] = encoder.fit_transform(df[cols])
```

```python
# dropping unnecessary columns
df.drop(['disp_size','bluetooth','mp_speed','dual_sim','num_cores','int_memory','ram','battery_power','mob_width','mob_height','mob_depth','mob_weight'],axis = 1,inplace = True)
```

```python
x = df.drop('mobile_price',axis = 1)
y = df['mobile_price']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
knr = KNeighborsRegressor()
knr.fit(x_train,y_train)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
# for linear model                                             
y_pred1 = lr.predict(x_test)
mse1 = mean_squared_error(y_test,y_pred1)
mse1
```

```python
#  for KNearestNeighbor
y_pred2 = knr.predict(x_test)
mse2 = mean_squared_error(y_test,y_pred2)
mse2
```

```python
# for DecisionTreeRegressor
y_pred3 = dtr.predict(x_test)
mse3 = mean_squared_error(y_test,y_pred3)
mse3
```
